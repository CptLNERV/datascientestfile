```python
from pyspark.sql import SparkSession


spark = SparkSession.builder.appName("Pipelines Spark ML").getOrCreate()
```

```python
import os
pwd = os.getcwd()
print(pwd)
print("{}/data/miserables_full.txt".format(pwd))
```

```python
hr = spark.read.csv("{}/data/HR_comma_sep.csv".format(pwd),header=True)
hr.sample(False,0.001,seed = 111).toPandas()
```

```python
hr.printSchema()
hr = hr.select(
        hr.left,
        hr.satisfaction_level,
        hr.last_evaluation,
        hr.number_project,
        hr.average_montly_hours,
        hr.time_spend_company,
        hr.Work_accident,
        hr.promotion_last_5years,
        hr.sales,
        hr.salary
)

hr.printSchema()

print(hr.summary().toPandas())

print(hr.describe().toPandas())
```

```python
from pyspark.ml.feature import StringIndexer

SalesIndexer = StringIndexer(inputCol = "sales",outputCol = "indexdsales").fit(hr)
print(type(SalesIndexer))

hrSalesIndexed = SalesIndexer.transform(hr)


hrSalesIndexed.sample(False,0.001,seed=111).toPandas()
```

```python
from pyspark.ml.feature import IndexToString

salesReconstructed = IndexToString(inputCol="indexdsales",outputCol="salesReconstructed",labels=SalesIndexer.labels)

hrSalesReconstructed = salesReconstructed.transform(hrSalesIndexed)

hrSalesReconstructed.sample(False,0.001,seed=222).toPandas()
```

```python
from pyspark.ml import Pipeline

SalesIndexer = StringIndexer(inputCol="sales",outputCol ="indexedSales")
SalaryIndexer = StringIndexer(inputCol="salary",outputCol ="indexedSalary")

indexer = Pipeline(stages=[SalesIndexer,SalaryIndexer])

hrIndexed = indexer.fit(hr).transform(hr)

hrIndexed.sample(False,0.001,seed = 111).toPandas()
```

```python
from pyspark.ml.linalg import DenseVector

hrNumeric = hrIndexed.select(
        hrIndexed.left,
        hrIndexed.satisfaction_level,
        hrIndexed.last_evaluation,
        hrIndexed.number_project,
        hrIndexed.average_montly_hours,
        hrIndexed.time_spend_company,
        hrIndexed.Work_accident,
        hrIndexed.promotion_last_5years,
        hrIndexed.indexedSales,
        hrIndexed.indexedSalary
)

hrNumeric.show()

hrRdd = hrNumeric.rdd.map(lambda x:(x[0],DenseVector(x[1:])))
hrLibsvm = spark.createDataFrame(hrRdd,["label","features"])

hrLibsvm.sample(False,0.001,seed=333).toPandas()
```

```python
hrNumeric.describe().toPandas()
```

```python
from pyspark.ml.feature import VectorIndexer

featureIndexer = VectorIndexer(inputCol="features",
                               outputCol="indexedFeatures",
                               maxCategories=10).fit(hrLibsvm)
```

```python
from pyspark.ml.classification import RandomForestClassifier

hr.select("Sales").distinct().show()

labelIndexer = StringIndexer(inputCol = "label",outputCol="indexedlabel").fit(hrLibsvm)

featureIndexer = VectorIndexer(inputCol="features",outputCol="indexedFeatures",maxCategories=10).fit(hrLibsvm)

rf = RandomForestClassifier(labelCol="indexedlabel",featuresCol = "indexedFeatures",predictionCol="prediction",seed=111)

labelConverter =  IndexToString(inputCol="indexedlabel",outputCol="predictedLabel",labels=labelIndexer.labels)

pipeline = Pipeline(stages=[labelIndexer,featureIndexer,rf,labelConverter])

(train,test) = hrLibsvm.randomSplit([.7,.3],seed  = 777)

model = pipeline.fit(train)
```

```python
predictions = model.transform(test)

predictions.sample(False,.001,seed=111).toPandas()
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Cration d'un valuateur 
evaluator = MulticlassClassificationEvaluator(metricName='accuracy',
                                              labelCol= 'indexedlabel',
                                              predictionCol= 'prediction')
# Calcul et affichage de la prcision du modle 
accuracy = evaluator.evaluate(predictions)
print(accuracy)
```
