```python
from pyspark.sql import SparkSession
from pyspark import SparkContext

sc = SparkContext.getOrCreate()

spark = SparkSession.builder.appName("Introduction à Spark ML").getOrCreate()

spark
```

```python
import sys
sys.path.append('/home/jayl/.local/lib/python3.10/site-packages')

import os
pwd = os.getcwd()
print(pwd)
print("{}/data/miserables_full.txt".format(pwd))
```

```python
# df_raw = sc.textFile("{}/data/YearPredictionMSD.txt".format(pwd))
# df_raw.take(5)

df_raw = spark.read.csv("{}/data/YearPredictionMSD.txt".format(pwd))
df_raw.show(5)

df_raw.sample(fraction = 0.001,seed=111).toPandas().head(5)
```

```python
df_raw.printSchema()
```

```python
from pyspark.sql.functions import col

exprs = [col(c).cast("double") for c in df_raw.columns]
df = df_raw.select(*exprs)
df.show(5)
df.printSchema()
```

```python
help(df.describe)
```

```python
df.describe().toPandas()
```

```python
# !pip install numpy
```

```python
from pyspark.ml.linalg import DenseVector

rdd_ml = df.rdd.map(lambda x:(x[0],DenseVector(x[1:])))
rdd_ml.take(5)

df_ml = spark.createDataFrame(rdd_ml,["label","features"])
df_ml.show(5)
```

```python
train,test=df_ml.randomSplit([0.7,0.3],seed=111)
```

```python
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(labelCol="label",featuresCol="features")

linearModel = lr.fit(train)

linearModel
```

```python
predicted = linearModel.transform(test)
predicted.show(5)
```

```python
print("RMSE",linearModel.summary.rootMeanSquaredError)
```

```python
print("R2",linearModel.summary.r2)
```

```python
from pprint import pprint 
pprint(linearModel.coefficients)
```
